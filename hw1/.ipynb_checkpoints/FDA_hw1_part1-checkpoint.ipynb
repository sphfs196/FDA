```python
import pandas as pd
from pandas import Series, DataFrame
filePath = './Reviews.csv'
df = pd.read_csv(filePath)  
df = df[:10000]
```

```python
### question 1
df['score count'] = Series(0, index=df.index)
for i, row in df.iterrows():
    df.at[i, 'score count'] = len(df[df['UserId'] == row['UserId']])
```

```python
df_sorted = df.sort_values(["score count", "UserId"], ascending=[False, True])
df_sorted.head(70)
```

```python
topDown_id = []
topDown_name = []
topDown_count = []
topDown_mean = []
cur_score = 0
for i, row in enumerate(df_sorted.iterrows()):
    index, data = row
    
    if i == 0:
        topDown_id.append(data['UserId'])
        topDown_name.append(data['ProfileName'])
        topDown_count.append(data['score count'])
        cur_score = cur_score + data['Score']
        
    elif data['UserId'] == topDown_id[-1] :         
        cur_score = cur_score + data['Score']        
        
    elif data['UserId'] != topDown_id[-1] :
        topDown_mean.append(cur_score/topDown_count[-1])
        topDown_id.append(data['UserId'])
        topDown_name.append(data['ProfileName'])
        topDown_count.append(data['score count'])
        
        cur_score = data['Score']
        
topDown_mean.append(cur_score/topDown_count[-1])
```

```python
print('UserId         ', ' profileName'.ljust(50), 'score count'.rjust(10), '   score mean'.rjust(10))
for i in range(len(topDown_count)):
    print(str(topDown_id[i]), ' ', str(topDown_name[i]).ljust(45), ' ', str(topDown_count[i]).rjust(10), ' ', str(topDown_mean[i]).rjust(10))
    if i >= 9:
        break
```

```python
### question 3
import time
df_sorted['year'] = Series(0, index=df_sorted.index)
for i, row in enumerate(df_sorted.iterrows()):
    index, data = row
    timeStamp = int(data['Time'])
    timeArray = time.localtime(timeStamp)
    otherStyleTime = time.strftime("%Y-%m-%d %H:%M:%S", timeArray)
    df_sorted.at[i, 'year'] = otherStyleTime[0:4]
    
```

```python
year_max = int(df_sorted.loc[df_sorted['year'].idxmax()]['year'])
year_min = int(df_sorted.loc[df_sorted['year'].idxmin()]['year'])
```

```python
years = np.arange(year_max - year_min + 1)+2000
years
```

```python
year_count = []
for y in years:
    year_count.append(len(df_sorted[df_sorted['year'] == y]))
year_count
```

```python
plt.bar(years, year_count)
```

```python
### question 5
ratio = []
for i, row in enumerate(df_sorted.iterrows()):
    #print(i)
    index, data = row
    if data['HelpfulnessNumerator'] > data['HelpfulnessDenominator']:
        continue
    elif data['HelpfulnessNumerator'] == 0 and data['HelpfulnessDenominator'] == 0:
        ratio.append(-1)
    else:
        ratio.append(data['HelpfulnessNumerator'] / data['HelpfulnessDenominator'])
```
